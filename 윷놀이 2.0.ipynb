```python
import random
import time

point = random.randint(1,5)

my_score = 0
enemy_score = 0


while  my_score <30 and enemy_score <30:
    
    while True:#내 턴
        print('>>>당신이 윷을 던졌습니다!')
        point = random.randint(1,5)
        
        for i in range(4):
            time.sleep(0.5)
            print('.',end="")
            if i != 3:
                
                continue
            else:
                print()
                break
                
        time.sleep(0.5)
        print()
 
        if point == 1:
            my_score+=1
            
            print("도! 1점을 획득하여 현재 당신의 점수는 {0}점입니다.".format(my_score))
            if my_score <30:
                print("다음은 상대방의 차례입니다.")
                break
            
            break
        elif point == 2:
            my_score+=2
            
            print("개! 2점을 획득하여 현재 당신의 점수는 {0}점입니다.".format(my_score))
            if my_score <30:
                print("다음은 상대방의 차례입니다.")
                break
            
            break
        elif point == 3:
            my_score+=3
            
            print("걸! 3점을 획득하여 현재 당신의 점수는 {0}점입니다.".format(my_score))
            if my_score < 30:
                print("다음은 상대방의 차례입니다.")
                break
            
            break
        elif point == 4:
            my_score+=4
            
            print("윷! 4점을 획득하여 현재 당신의 점수는 {0}점입니다.".format(my_score))
            if my_score <30:
                print("당신에게 한번 더 던질 기회가 주어집니다.")
                continue
            
            break
        elif point == 5:
            my_score+=5
            
            print("모! 5점을 획득하여 현재 당신의 점수는 {0}점입니다.".format(my_score))
            if my_score <30:
                print("당신에게 한번 더 던질 기회가 주어집니다.")
                continue
            
            break
    
    if my_score >=30:
        break
        
    while True:#상대방 턴
        print('>>>상대방이 윷을 던졌습니다!')
        point = random.randint(1,5)
        
        for i in range(4):
            time.sleep(0.5)
            print('.',end="")
            if i != 3:
                continue
            else:
                print()
                break
                
        time.sleep(0.5)
        print()
    
        if point == 1:
            enemy_score+=1
            
            print("도! 1점을 획득하여 현재 상대방의 점수는 {0}점입니다.".format(enemy_score))
            if enemy_score < 30:
                print("다음은 당신의 차례입니다.")
                break
            
            break
        elif point == 2:
            enemy_score+=2
            
            print("개! 2점을 획득하여 현재 상대방의 점수는 {0}점입니다.".format(enemy_score))
            if enemy_score < 30:
                print("다음은 당신의 차례입니다.")
                break
            
            
            break
        elif point == 3:
            enemy_score+=3
            
            print("걸! 3점을 획득하여 현재 상대방의 점수는 {0}점입니다.".format(enemy_score))
            if enemy_score < 30:
                print("다음은 당신의 차례입니다.")
                break
            
            break
        elif point == 4:
            enemy_score+=4
            
            print("윷! 4점을 획득하여 현재 상대방의 점수는 {0}점입니다.".format(enemy_score))
            if enemy_score < 30:
                print('상대방에게 한번 더 던질 기회가 주어집니다.')
                continue
            
            break
        elif point == 5:
            enemy_score+=5
            
            print("모! 5점을 획득하여 현재 상대방의 점수는 {0}점입니다.".format(enemy_score))
            if enemy_score < 30:  
                print('상대방에게 한번 더 던질 기회가 주어집니다.')
                continue
            
            break
    
    if enemy_score >=30:
        break
            
        

if (my_score >=30):
    print(">>>당신이 이겼습니다. 당신의 점수: {0}, 상대방의 점수: {1}".format(my_score,enemy_score))
    
else:
    print(">>>상대방이 이겼습니다. 상대방의 점수: {0}, 당신의 점수: {1}".format(enemy_score,my_score))
```
